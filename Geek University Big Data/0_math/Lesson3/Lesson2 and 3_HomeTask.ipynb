## Эллипс

```python
x = np.linspace(-100,100,100000)
a = 100
b = 9
y1 = np.sqrt((1 - x**2/a**2)*b**2)
y2 = -np.sqrt((1 - x**2/a**2)*b**2)
```

```python
plt.figure(figsize=[40,8])
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

## Гипербола

```python
x = np.linspace(-5,5,100000)
a = 2
b = 0.3
y1 = np.sqrt((1 - x**2/a**2)*(-b**2))
y2 = -np.sqrt((1 - x**2/a**2)*(-b**2))
```

```python
plt.figure(figsize=[25,8])
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

#### 4. Задание (в программе)
Решите систему уравнений:
- y = x2 – 1
- exp(x) + x∙(1 – y) = 1



```python
from scipy.optimize import fsolve
```

```python
def xx(p):
    x,y = p
    return(y-x**2+1, np.exp(x) + x*(1-y)-1)
```

```python
(x1,x2) = fsolve(xx,(1,1))
```

```python
print(x1,x2)
```
